```python
%%writefile submit.py
String = sorted(input())
String = ''.join(String)
alpha='abcdefghijklmnopqrstuvwxyz'
n=len(String)
array=[]
num_rep=[]
check=0

for p in alpha:
    if String.count(p) >1:
        check = 1
        num_rep.append(String.count(p))


for j in range(n):
    array.append(j)

    
def RotateR(array,k,i):
    temp = array[i]
    for t in range(i-k): 
        array[i-t] = array[i-t-1]
    array[k] = temp

def RotateL(array,k,i):
    temp = array[k]
    for t in range(i-k): 
        array[k+t] = array[k+t+1]
    array[i] = temp


list_a=[0]

def per(k):
    if k == n:
        a=''
        for k in array:
            a=a+String[k]
        #print(a)
        ch=0
        for t in list_a:
            if a == t :
                ch=1
        
        if ch==0 :
            print(a)
            list_a.append(a)
            
    else:
        for i in range(k,n):
            RotateR(array,k,i)
            per(k+1)
            RotateL(array,k,i)

def per_no_rep(k):
    if k == n:
        a=''
        for k in array:
            a=a+String[k]
        print(a)
            
    else:
        for i in range(k,n):
            RotateR(array,k,i)
            per_no_rep(k+1)
            RotateL(array,k,i)
            
if check == 1:            
    fac=1
    for c in range(n):
        fac *= (c+1)
    
    rep=1
    for q in num_rep:
        for b in range(q):
            rep*=(b+1)
    print(int(fac/rep))
    per(0)
else:
    fac=1
    for c in range(n):
        fac *= (c+1)
    print(fac)
    per_no_rep(0)
```
